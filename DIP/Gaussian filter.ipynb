```python
#Gaussian filter
import numpy as np
import cv2
from matplotlib import pyplot as plt
from PIL import Image, ImageFilter
from math import log10,sqrt
%matplotlib inline
image = cv2.imread('sam.JPG')
new_image = cv2.GaussianBlur(image, (3, 3),1)
plt.figure(figsize=(14,16))
plt.subplot(121), plt.imshow(cv2.cvtColor(image,cv2.COLOR_HSV2RGB)),plt.title('Original image')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(cv2.cvtColor(new_image,cv2.COLOR_HSV2RGB)),plt.title('gaussian filter')
plt.xticks([]), plt.yticks([])
plt.show()
mse = np.mean((image - new_image) ** 2)
if(mse == 0): # MSE is zero means no noise is present in the signal .
# Therefore PSNR have no importance.
    print(100)
else:
    max_pixel = 255.0
    psnr = 20 * log10(max_pixel / sqrt(mse))
    print(psnr)
```
